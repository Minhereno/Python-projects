```python
import pandas as pd 
import numpy as np
import plotly.express as px
```

# SARS-CoV-2 Testing Metrics Among Health Districts in VA (2022)




```python
#Read and clean data
df = pd.read_csv('https://data.virginia.gov/api/views/3u5k-c2gr/rows.csv?accessType=DOWNLOAD') #read csv
df['Lab Report Date'] = pd.to_datetime(df['Lab Report Date'], errors = 'coerce').dt.date
df['PCR Positivity Rate (%)'] = round((df['Number of Positive PCR Testing Encounters']/df['Number of PCR Testing Encounters'])*100, 2) #positivity rate calc
df = df[df['Lab Report Date'] >= pd.to_datetime('2022')] #Only get lab report date in 2022
df = df[df['Health District'] != 'Out of State'] #Remove out of state health district
df = df[df['Health District'] != 'Unknown'] #Remove unknown health districts
df
```

```python
#Functions for 5% PCR and 10% PCR calculations
def posfunc(df): 
    if df['PCR Positivity Rate (%)'] == 0:
        return 'Null'
    elif df['PCR Positivity Rate (%)'] <= 5:
        return 'Yes'
    else:
        return 'No'

def posfunc2(df): 
    if df['PCR Positivity Rate (%)'] == 0:
        return 'Null'
    elif df['PCR Positivity Rate (%)'] <= 10:
        return 'Yes'
    else:
        return 'No'
        
#Applying these calculations to dataframe
df['5% PCR Positive Testing Rate'] = df.apply(posfunc,axis=1)
df['10% PCR Postitive Testing Rate'] = df.apply(posfunc2,axis=1)
```

```python
df.groupby(['Health District'])['Health District'].count() #getting counts of health districts
```

```python
df.groupby(['Health District']).sum()
```

### Mean PCR Positivity Rate

```python
#Compare mean PCR positivity rate over different health districts in Virginia (2022)
mean_pcr_df = df.groupby(['Health District'])['PCR Positivity Rate (%)'].mean().reset_index() #groupby health district and calculate mean of PCR positivity rate
mean_pcr_df = mean_pcr_df.rename(columns={'PCR Positivity Rate (%)':'Mean PCR Positivity Rate (%)'}) #rename column to mean PCR positivity rate
mean_pcr_df = mean_pcr_df.sort_values('Mean PCR Positivity Rate (%)') #sort values by mean PCR positivity rate
mean_pcr_df
```

```python
mean_pcr_fig = px.bar(mean_pcr_df, x = 'Health District', y = 'Mean PCR Positivity Rate (%)', title = 'Mean PCR Positivity Rate of Health Districts of Virginia in 2022') #barplot
mean_pcr_fig.add_hline(y=5, line_color = 'red') #add a 5% posivity rate reference line
```

### Total PCR Tests

```python
num_pcr_test_df = df.groupby(['Health District'])['Number of PCR Testing Encounters'].sum().reset_index()
num_pcr_test_df = num_pcr_test_df.sort_values('Number of PCR Testing Encounters')
num_pcr_test_df
```

```python
num_pcr_test_fig = px.bar(num_pcr_test_df, x = 'Health District', y = 'Number of PCR Testing Encounters', title = 'Total Number of PCR Testing Among Health Districts in Virginia in 2022') #barplot
num_pcr_test_fig
```

### Total Number of Testing

```python
num_total_test_df = df.groupby(['Health District'])['Total Number of Testing Encounters'].sum().reset_index()
num_total_test_df = num_total_test_df.sort_values('Total Number of Testing Encounters')
num_total_test_df
```

```python
num_pcr_test_fig = px.bar(num_total_test_df, x = 'Health District', y = 'Total Number of Testing Encounters', title = 'Total Number of Testing Among Health Districts in Virginia in 2022') #barplot
num_pcr_test_fig
```
